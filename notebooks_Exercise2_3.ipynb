```python
#importing the needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#creating the data set
num_samples = 1000
num_ft = 2

a = [10,5]*np.random.randn(num_samples, num_ft) + [150,50]
target_1 = np.zeros(1000)

b = [10,5]*np.random.randn(num_samples, num_ft) + [50,25]
target_2 = np.ones(1000)

#concatenating the data

Dataset = np.concatenate((a, b))
Target = np.concatenate((target_1, target_2))
print(Dataset.shape)
print(Target.shape)
```

```python
#scatter plot
plt.scatter(Dataset[:,0], Dataset[:,1],c=Target)
plt.title("Dataset graph")
plt.xlabel("Height (cm)")
plt.ylabel("Weight(kg)")
plt.show()
```

```python
#saving the dataset into a csv file
Dataset2 =pd.DataFrame(Dataset)
Dataset2.to_csv('Dataset.csv')
Dataset3= pd.read_csv('Dataset.csv',header=None)
#print(Dataset2)

#saving the target into a csv file
Target2 =pd.DataFrame(Target)
Target2.to_csv('Target.csv')
Target3= pd.read_csv('Target.csv',header=None)
#print(Target2)
```
